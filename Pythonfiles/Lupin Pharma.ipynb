```python
from tabula import read_pdf
from tabulate import tabulate
import PyPDF2 as pdf
import re
import camelot
```

```python
# open the pdf file
object = pdf.PdfFileReader("lupin-annual-report-2018-19.pdf")

# get number of pages
NumPages = object.getNumPages()

# define keyterms
String = "Consolidated Balance Sheet"
String1 = "as at March 31, 2019"

# extract text and do the search
for i in range(0, NumPages):
    PageObj = object.getPage(i)
    Text = PageObj.extractText() 
    # print(Text)
    if  re.search(String, Text) != None and re.search(String1, Text) != None:
        print("This is page " + str(i))
        page=object.getPage(i)
        pdf_writer=pdf.PdfFileWriter()
        pdf_writer.addPage(page)
        output=open('balancesheet_1819.pdf','wb')
        pdf_writer.write(output)
        output.close()
        df=read_pdf('balancesheet_1819.pdf')
df[0]
```

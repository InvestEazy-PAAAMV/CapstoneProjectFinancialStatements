# Initialization of Packages

```python
from tabula import read_pdf
from tabulate import tabulate
import PyPDF2 as pdf
import re
```

# For 2018-19

## Consolidated Balance Sheet

```python
# open the pdf file
object = pdf.PdfFileReader("APL-Annual-Report-2018-19.pdf")

# get number of pages
NumPages = object.getNumPages()

# define keyterms
String = "CONSOLIDATED BALANCE SHEET"

# extract text and do the search
for i in range(0, NumPages):
    PageObj = object.getPage(i)
    Text = PageObj.extractText() 
    # print(Text)
    if  re.search(String, Text) != None:
        print("This is page " + str(i))
        page=object.getPage(i)
        pdf_writer=pdf.PdfFileWriter()
        pdf_writer.addPage(page)
        output=open('balancesheet_aurobindo_1819.pdf','wb')
        pdf_writer.write(output)
        output.close()
        df=read_pdf('balancesheet_aurobindo_1819.pdf')
df[0]
```

```python
#Chairmans Message
# open the pdf file
object = pdf.PdfFileReader("APL-Annual-Report-2018-19.pdf")

# get number of pages
NumPages = object.getNumPages()

# define keyterms
String = "VICE CHAIRMAN’S REVIEW"
String1 = "Vice Chairman"

# extract text and do the search
for i in range(0, NumPages):
    PageObj = object.getPage(i)
    Text = PageObj.extractText() 
    # print(Text)
    if  re.search(String, Text) != None and re.search(String1, Text)!=None:
        print("This is page " + str(i))
        page=object.getPage(i)
        pdf_writer=pdf.PdfFileWriter()
        pdf_writer.addPage(page)
        output=open('cmessage_1819.pdf','wb')
        pdf_writer.write(output)
        output.close()
```

## Consolidated Statement of Cash Flows

```python
# open the pdf file
object = pdf.PdfFileReader("Annual-Report-2016-17.pdf")

# get number of pages
NumPages = object.getNumPages()
String1 = "Consolidated Cash Flow Statement for the year ended March 31, 2017"
String2 = "B. Adi Reddy"
String3 = "ICAI Firm Registration No. 101049W/E300004"
String4 = "CASH AND CASH EQUIVALENTS AT THE END OF THE YEAR"
pdf_writer=pdf.PdfFileWriter()
for i in range(0, NumPages):
    PageObj = object.getPage(i)
    Text = PageObj.extractText() 
    # print(Text)
    if  re.search(String1, Text) != None or (re.search(String2, Text) != None and re.search(String3, Text) != None) and re.search(String4, Text) != None: 
        print("This is page " + str(i))
        page=object.getPage(i)
        pdf_writer.addPage(page)
output=open('cashflow_1617.pdf','wb')
pdf_writer.write(output)
output.close()
```
